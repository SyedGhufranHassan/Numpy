You are working on a project that involves image processing. In this assignment, you will utilize NumPy to implement various operations on an RGB image of your own choice. Follow the steps below:

- Select an Image:
- Choose an RGB image of your own choice. You can use any image. Ensure that the image is in RGB format.
- Convert RGB Image to Grayscale:
- Convert the grayscale image to a binary image using a threshold value.
- Apply Gaussian blur to the binary image to reduce noise.
- Detect edges in the blurred image using the Sobel operator.
- Compute the histogram of pixel intensities in the original grayscale image.


<br> Visualize the results of each operation to observe the effects on the image.
Submission Guidelines:

<br>Submit the selected RGB image in the submission as well.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


image = cv2.imread('pakmap.jpg')

rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
rgb_image



```

```python

gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
gray_image
```

```python
# Convert RGB image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
```

```python
# Convert RGB image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
```

```python
# Convert grayscale image to binary image using a threshold
_, binary_image = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY)
```

```python


# Apply Gaussian blur to the binary image
blurred_image = cv2.GaussianBlur(binary_image, (5, 5), 0)
```

```python
# Detect edges in the blurred image using the Sobel operator
sobel_x = cv2.Sobel(blurred_image, cv2.CV_64F, 1, 0, ksize=5)
sobel_y = cv2.Sobel(blurred_image, cv2.CV_64F, 0, 1, ksize=5)
edges_image = np.sqrt(sobel_x**2 + sobel_y**2)
```

```python


# Compute histogram of pixel intensities in the original grayscale image
histogram = np.histogram(gray_image, bins=256, range=(0, 255))
```

```python


# Visualize the results
plt.figure(figsize=(12, 6))

plt.subplot(2, 3, 1)
plt.imshow(rgb_image)
plt.title('Original RGB Image')

plt.subplot(2, 3, 2)
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')

plt.subplot(2, 3, 3)
plt.imshow(binary_image, cmap='gray')
plt.title('Binary Image')

plt.subplot(2, 3, 4)
plt.imshow(blurred_image, cmap='gray')
plt.title('Blurred Image')

plt.subplot(2, 3, 5)
plt.imshow(edges_image, cmap='gray')
plt.title('Edges Image')

plt.subplot(2, 3, 6)
plt.plot(histogram[1][:-1], histogram[0], color='black')
plt.title('Histogram of Grayscale Image')

plt.tight_layout()
plt.show()
```
